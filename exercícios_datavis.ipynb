```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Conhecendo a biblioteca Matplotlib

```python
url = "https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv"
dados = pd.read_csv(url)
dados
```

Neste momento temos uma nova demanda: criar um gráfico de linhas comparando os números de imigrantes do Brasil e Argentina para o Canadá, que são os maiores países da América do Sul

```python
dados.set_index('País', inplace = True)
anos =  list(map(str, range(1980, 2014)))
```

```python
brasil_argentina = dados.loc[['Brasil', 'Argentina'], anos]
brasil_argentina
```

```python
brasil_argentina = dados.loc[['Brasil', 'Argentina'], anos]
brasil_argentina = brasil_argentina.transpose()
brasil_argentina_dict = {'ano':brasil_argentina.index.tolist() , 'Brasil': brasil_argentina['Brasil'], 'Argentina': brasil_argentina['Argentina']}
dados_brasil_argentina = pd.DataFrame(brasil_argentina_dict)
dados_brasil_argentina.head()
```

```python
plt.plot(dados_brasil_argentina['ano'], dados_brasil_argentina['Brasil'], dados_brasil_argentina['Argentina'])
plt.title('Imigrantes do Brasil e da Argentina para o Canadá')
plt.xlabel('anos')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend(['Brasil', 'Argentina'])
plt.show()
```

Resolução da Alura:

```python
plt.plot(brasil_argentina['Brasil'],label = 'Brasil', color = 'g')
plt.plot(brasil_argentina['Argentina'],label ='Argentina', color='m')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

## 2. Criando figuras com Matplotlib

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
vendas_por_loja = pd.DataFrame(vendas_2022, index=lojas)
vendas_por_loja.head()
```

```python
 mes = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
```

```python
# Criar a figura e os subplots:
fig, axs = plt.subplots(2,2, figsize=(12,8))
# Ajustar os espaçamentos entre os subplots:
fig.subplots_adjust(hspace=0.4, wspace=0.3)
# Título geral:
fig.suptitle('Quantidade de Vendas de cada loja no ano de 2022', fontsize = 16)


# Gráficos de cada loja:
axs[0,0].plot(vendas_por_loja.loc['A', mes], color = 'crimson')
axs[0,0].set_title('Vendas da Loja A', loc = 'left')

axs[0,1].plot(vendas_por_loja.loc['B', mes], color = 'gold')
axs[0,1].set_title('Vendas da Loja B', loc =  'left')

axs[1,0].plot(vendas_por_loja.loc['C', mes], color = 'slateblue')
axs[1,0].set_title('Vendas da Loja C', loc = 'left')

axs[1,1].plot(vendas_por_loja.loc['D', mes], color = 'darkturquoise')
axs[1,1].set_title('Vendas da Loja D', loc = 'left')


# Legendas para cada um dos eixos:
for ax in axs.flat:
  ax.set_xlabel('Mês')
  ax.set_ylabel(' de Vendas')


# Padronizar a escala do eixo Y:
ymin = 50
ymax = 450

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)


# Retirar demarcações que poluem o gráfico:
for ax in axs.flat:
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)

for i, ax in enumerate(axs.flat):
  ax.grid(color='lightgrey')

# Salvar o Gráfico:
fig.savefig('numero_de_vendas_por_loja_2022.png', transparent = False, dpi = 300, bbox_inches = 'tight')


plt.show()
```
